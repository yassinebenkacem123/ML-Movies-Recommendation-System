```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
movies = pd.read_csv('./tmdb_5000_movies.csv')
credits = pd.read_csv('./tmdb_5000_credits.csv')
```

```python
movies.head(3)
```

```python
credits.head(3)
```

```python
print("movies shpape:", movies.shape)
print("credits shpape:", credits.shape)
```

```python
movies = movies.merge(credits, on ='title')
movies.shape
```

```python
movies.columns
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(3)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
type(movies[['genres']].iloc[0,0])
# so we need to convert the string representation of list of dictionaries to actual list of dictionaries
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
type(movies[['genres']].iloc[0,0])
```

```python
movies.head(3)
```

```python
# convert the 'cast' column to a list of top 3 actors
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(3)
```

```python
movies['overview'] = movies['overview'].str.split()
movies.head(3)
```

```python
movies['overview'][0]
```

```python
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head(3)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].str.lower()
```

```python
new_df['tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# this stem the words, and  that mean : removing suffixes like "ing", "ly", "es", "s" etc to get to the root form of the word
def stems (text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Spider')
```
